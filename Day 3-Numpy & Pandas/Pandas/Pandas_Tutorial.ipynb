# crosstab()
This method is used to compute a simple cross-tabulation of two (or more) factors. By default, computes a frequency table of the factors unless an array of values and an aggregation function are passed.

```python
pd.crosstab( titanic_data.sex, titanic_data.survived )
```

How to Normalize a Pandas DataFrame Row?

```python
pd.crosstab( titanic_data.sex, titanic_data.survived, normalize = "index" )
```

How to Normalize a Pandas DataFrame Column?

```python
pd.crosstab( titanic_data.sex, titanic_data.survived, normalize = "columns" )
```

How to Normalize a Pandas DataFrame?

```python
pd.crosstab( titanic_data.sex, titanic_data.survived, normalize = "all" )
```

# sample()
Pandas sample() is used to generate a sample random row or column from the function caller data frame.

Syntax:

DataFrame.sample(n=None, frac=None, replace=False, weights=None, random_state=None, axis=None)

```python
titanic_data.sample(8)
```

```python
# Fill missing values in the 'age' column with the mean age
titanic_data['age'].fillna(titanic_data['age'].mean(), inplace=True)

# Fill missing values in other columns with default values
titanic_data['cabin'].fillna('Unknown', inplace=True)
titanic_data['embarked'].fillna('Unknown', inplace=True)
titanic_data['home.dest'].fillna('Unknown', inplace=True)
```

```python
titanic_data.sample(8)
```

```python
# Convert the 'sex' column to numeric values
titanic_data['sex'] = titanic_data['sex'].map({'female': 0, 'male': 1})
```

```python
titanic_data.sample(8)
```

```python
# Remove unnecessary columns
titanic_data.drop(['boat', 'body'], axis=1, inplace=True)
```

```python
titanic_data.sample(8)
```

```python
# Improve the formatting of names
titanic_data['name'] = titanic_data['name'].str.title()
```

```python
# Convert the 'fare' column to float type
titanic_data['fare'] = titanic_data['fare'].astype(float)
```

```python
titanic_data.sample(8)
```

```python
# Check for duplicates
print(titanic_data.duplicated().sum())

# Remove duplicates
titanic_data.drop_duplicates(inplace=True)
```

```python
# Create age groups
bins = [0, 18, 35, 50, 100]
labels = ['Child', 'Young Adult', 'Adult', 'Senior']
titanic_data['age_group'] = pd.cut(titanic_data['age'], bins=bins, labels=labels)

# Create family size feature
titanic_data['family_size'] = titanic_data['sibsp'] + titanic_data['parch'] + 1
```

```python
titanic_data.sample(8)
```

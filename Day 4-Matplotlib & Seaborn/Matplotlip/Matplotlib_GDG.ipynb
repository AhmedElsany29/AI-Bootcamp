# Introduction to Matplotlib
- **Matplotlib** is a powerful library in Python for creating a wide variety of visualizations. It's widely used in data analysis and scientific research for plotting data effectively.

## Key Features of Matplotlib
- Open-source library.
- Provides the pyplot module for simple plotting.
- Supports various plot types: line plots, bar plots, scatter plots, pie charts, 3D plots, and more.
- Highly customizable for professional-quality visualizations.

### How to Choose the Right Visualization

- **Numerical Data**
    - Univariate
        - Histograms
        - Density Plot
    - Bivariate / Multivariate
        - Scatter Plots
        - Histograms
        - Boxplots
        - Line Plots
        - Density Plots
        - Area Plots
        - Violin Plots
- **Categorical Data**
    - Univariate
        - Barplots
        - Pie Charts
        - Words Clouds
    - Bivariate / Multivariate
        - Barplots
        - Heatmaps
        - Treemaps
        - Dendrograms

## Numerical Data


### 1-Univariate Visualization

#### 1-Histogram:

- Useful for showing the distribution of numerical data.
- Represents the frequency of values within intervals (bins).

```python
data = np.random.randn(1000)

plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
data = np.random.randn(1000)

plt.hist(data, bins=20, color='red', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

#### 2-Density Plot (approximated with plt.plot):

```python
data = np.random.normal(size=1000)
sorted_data = np.sort(data)
density = (1 / (np.sqrt(2 * np.pi))) * np.exp(-0.5 * sorted_data**2)

plt.plot(sorted_data, density, color="purple")
plt.title("Density Plot Example")
plt.xlabel("Values")
plt.ylabel("Density")
plt.show()
```

### 2-Bivariate / Multivariate Visualization


#### 1-Scatter Plot:

- Displays the relationship between two numerical variables.

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y, color='red', marker='o')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

#### Difference Between Marker & Attribute:

- Marker: Controls the shape of points, **e.g., o, s, ^.**
- Attribute: Includes broader styling options like **color, size, and label.**

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y, color='blue', marker='s')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y, color='green', marker='^')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Create a figure with a larger size to accommodate the labels
fig = plt.figure(figsize=(6, 4), dpi=150)  # Increased figure size

# Define data
subjects = ["Machine Learning", "Maths", "Deep Learning", "NLP"]
student1_marks = [92, 88, 60, 72]
student2_marks = [65, 78, 96, 85]

# Plot std1 and std2 against subjects
plt.scatter(subjects, student1_marks, label='Student 1')  # Scatter plot for Student 1
plt.scatter(subjects, student2_marks, label='Student 2')  # Scatter plot for Student 2

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')  # Rotate labels and align them to the right

# Add padding between the x-axis labels and the plot
plt.tight_layout()

# Add a legend to the plot
plt.legend()

# Display the plot
plt.show()
```

#### 2-Boxplot

- Summarizes the distribution of data using quartiles.



```python
data = [np.random.randn(100), np.random.randn(100),np.random.randn(100)]

plt.boxplot(data, vert=True, patch_artist=True)
plt.title('Boxplot')
plt.xlabel('Groups')
plt.ylabel('Values')
plt.show()
```

#### 3-Line Plot

- Used to show trends over time or relationships.


```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, linestyle='--', color='purple', marker='o')
plt.title('Line Plot with Markers')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x = range(10)
y = [xi**2 for xi in x]

plt.plot(x, y, marker="^", color="red")
plt.title("Line Plot Example")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
# Evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# Plot multiple lines using *args
plt.plot(t, t, 'r--',  # Red dashed line
         t, t**2, 'bo',  # Blue circles
         t, t**3, 'g^')  # Green triangles

''' Explanation:
- The first argument in each triplet is the x-axis data (t).
- The second argument is the y-axis data (t, t**2, t**3).
- The third argument specifies the line style or marker:
  - 'r--': Red dashed line
  - 'bo': Blue circles
  - 'g^': Green triangles
'''

# Display the plot
plt.show()
```

#### 4-Area Plot:


```python
x = range(10)
y = [xi**2 for xi in x]

plt.fill_between(x, y, color="skyblue", alpha=0.5)
plt.title("Area Plot Example")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

## Categorical Data


### Univariate Visualization


#### 1-Bar Plot:

- Displays the frequency or total values for categories.

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 30]

plt.bar(categories, values, color='orange')
plt.title('Barplot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [50, 10, 15, 60]

plt.bar(categories, values, color='blue')
plt.title('Barplot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid()
plt.show()
```

#### 2- Pie Chart

- Used to show percentage distribution among categories.

```python
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90, colors=['red', 'blue', 'green', 'purple'])
plt.title('Pie Chart')
plt.show()
```

#### 3-Heatmap :


```python
categories = ['Product A', 'Product B', 'Product C', 'Product D']
stores = ['Store 1', 'Store 2', 'Store 3', 'Store 4']

# Example sales data
sales = np.random.randint(50, 200, size=(4, 4))

plt.imshow(sales, cmap="coolwarm", interpolation="nearest")
plt.title("Heatmap of Sales by Store and Product")
plt.xticks(ticks=range(len(categories)), labels=categories)
plt.yticks(ticks=range(len(stores)), labels=stores)
plt.colorbar(label="Sales")
plt.xlabel("Products")
plt.ylabel("Stores")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MultipleLocator, FuncFormatter

np.random.seed(19680801)

X = np.linspace(0.5, 3.5, 100)
Y1 = 3+np.cos(X)
Y2 = 1+np.cos(1+X/0.75)/2
Y3 = np.random.uniform(Y1, Y2, len(X))

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1, aspect=1)


def minor_tick(x, pos):
    if not x % 1.0:
        return ""
    return "%.2f" % x

ax.xaxis.set_major_locator(MultipleLocator(1.000))
ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.yaxis.set_major_locator(MultipleLocator(1.000))
ax.yaxis.set_minor_locator(AutoMinorLocator(4))
ax.xaxis.set_minor_formatter(FuncFormatter(minor_tick))

ax.set_xlim(0, 4)
ax.set_ylim(0, 4)

ax.tick_params(which='major', width=1.0)
ax.tick_params(which='major', length=10)
ax.tick_params(which='minor', width=1.0, labelsize=10)
ax.tick_params(which='minor', length=5, labelsize=10, labelcolor='0.25')

ax.grid(linestyle="--", linewidth=0.5, color='.25', zorder=-10)

ax.plot(X, Y1, c=(0.25, 0.25, 1.00), lw=2, label="Blue signal", zorder=10)
ax.plot(X, Y2, c=(1.00, 0.25, 0.25), lw=2, label="Red signal")
ax.plot(X, Y3, linewidth=0,
        marker='o', markerfacecolor='w', markeredgecolor='k')

ax.set_title("Anatomy of a figure", fontsize=20, verticalalignment='bottom')
ax.set_xlabel("X axis label")
ax.set_ylabel("Y axis label")

ax.legend()


def circle(x, y, radius=0.15):
    from matplotlib.patches import Circle
    from matplotlib.patheffects import withStroke
    circle = Circle((x, y), radius, clip_on=False, zorder=10, linewidth=1,
                    edgecolor='black', facecolor=(0, 0, 0, .0125),
                    path_effects=[withStroke(linewidth=5, foreground='w')])
    ax.add_artist(circle)


def text(x, y, text):
    ax.text(x, y, text, backgroundcolor="white",
            ha='center', va='top', weight='bold', color='blue')


# Minor tick
circle(0.50, -0.10)
text(0.50, -0.32, "Minor tick label")

# Major tick
circle(-0.03, 4.00)
text(0.03, 3.80, "Major tick")

# Minor tick
circle(0.00, 3.50)
text(0.00, 3.30, "Minor tick")

# Major tick label
circle(-0.15, 3.00)
text(-0.15, 2.80, "Major tick label")

# X Label
circle(1.80, -0.27)
text(1.80, -0.45, "X axis label")

# Y Label
circle(-0.27, 1.80)
text(-0.27, 1.6, "Y axis label")

# Title
circle(1.60, 4.13)
text(1.60, 3.93, "Title")

# Blue plot
circle(1.75, 2.80)
text(1.75, 2.60, "Line\n(line plot)")

# Red plot
circle(1.20, 0.60)
text(1.20, 0.40, "Line\n(line plot)")

# Scatter plot
circle(3.20, 1.75)
text(3.20, 1.55, "Markers\n(scatter plot)")

# Grid
circle(3.00, 3.00)
text(3.00, 2.80, "Grid")

# Legend
circle(3.70, 3.80)
text(3.70, 3.60, "Legend")

# Axes
circle(0.5, 0.5)
text(0.5, 0.3, "Axes")

# Figure
circle(-0.3, 0.65)
text(-0.3, 0.45, "Figure")

color = 'blue'
ax.annotate('Spines', xy=(4.0, 0.35), xytext=(3.3, 0.5),
            weight='bold', color=color,
            arrowprops=dict(arrowstyle='->',
                            connectionstyle="arc3",
                            color=color))

ax.annotate('', xy=(3.15, 0.0), xytext=(3.45, 0.45),
            weight='bold', color=color,
            arrowprops=dict(arrowstyle='->',
                            connectionstyle="arc3",
                            color=color))

ax.text(4.0, -0.4, "Made with http://matplotlib.org",
        fontsize=10, ha="right", color='.5')

plt.show()
```
